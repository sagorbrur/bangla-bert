## Sentiment Analysis

```
# data prepare
import pandas as pd
df = pd.read_csv('sentiment_analysis.csv')
df = df.sample(frac=1)
df.head()
```

```
from sklearn.model_selection import train_test_split

train, eval = train_test_split(df,test_size=0.2)
```

```
train.head()
```

```
eval.head()
```

## Bangla BERT

```
import pandas as pd

from simpletransformers.classification import ClassificationModel


# Create a ClassificationModel
model = ClassificationModel("bert", "sagorsarker/bangla-bert-base")

# Train the model
model.train_model(train)

# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(eval)
```

```
print(result)
```

## Multi-lingual BERT

```
import pandas as pd

from simpletransformers.classification import ClassificationModel


# Create a ClassificationModel
model = ClassificationModel("bert", "bert-base-multilingual-uncased")

# Train the model
model.train_model(train)

# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(eval)
```

```
print(result)
```
